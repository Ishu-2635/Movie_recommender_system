```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies=movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.sample(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum
```

```python
import ast
def convert(ob):
    list = []
    for i in ast.literal_eval(ob):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords'][0]
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords'][0]
```

```python
movies['cast'][0]
```

```python
def convert1(obj):
    li = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 5:
            li.append(i['name'])
        else:
            break
    return li
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies.sample(5)
```

```python
def director(obj):
    li = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            li.append(i['name'])
            break
    return li
```

```python
movies['Director'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['Director'] = movies['Director'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
movies.head(
    
)
```

```python
movies.isnull()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['Director']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']] 
```

```python
new_df.sample()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    x= []
    for i in text.split():
        x.append(ps.stem(i))
    return " ".join(x)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][6]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words = 'english', max_features = 4500)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names_out().tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
similar.shape
```

```python
sorted(list(enumerate(similar[0])),reverse = True, key = lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similar[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
